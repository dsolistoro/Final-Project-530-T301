## Descriptive variables

```python
import pandas as pd
import numpy as np
from scipy.stats import mode, skew, kurtosis

def describe_variable(data, column):
    stats = {
        'Variable': column,
        'Mean': np.mean(data[column]),
        'Std Dev': np.std(data[column]),
        'Min': np.min(data[column]),
        'Max': np.max(data[column]),
        'Range': np.max(data[column]) - np.min(data[column]),
        'Q1': np.percentile(data[column], 25),
        'Q3': np.percentile(data[column], 75),
        'IQR': np.percentile(data[column], 75) - np.percentile(data[column], 25),
        'Skewness': skew(data[column]),
        'Kurtosis': kurtosis(data[column])
    }
    return stats

# List of variables to analyze
variables = ['Sleep Hours', 'Calories Burned', 'Age', 'Workout Duration (mins)']

# Calculate descriptive statistics for each variable
descriptive_stats = []
for var in variables:
    stats = describe_variable(df, var)
    descriptive_stats.append(stats)

# Convert to a DataFrame for better visualization
stats_df = pd.DataFrame(descriptive_stats)
print(stats_df)
```

## PMF Sleep hours
Filter the Data based on Gender
- Scenario 1: Sleep hours for Women.
- Scenario 2: Sleep hours for Men.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for each scenario
sleep_women = df[df['Gender'] == 'Female']['Sleep Hours']
sleep_men = df[df['Gender'] == 'Male']['Sleep Hours']

def calculate_pmf(data):
    value_counts = data.value_counts().sort_index()
    pmf = value_counts / len(data)
    return pmf

# Calculate PMF for each scenario
pmf_women = calculate_pmf(sleep_women)
pmf_men = calculate_pmf(sleep_men)

# Plot PMFs
plt.figure(figsize=(10, 6))
plt.bar(pmf_women.index, pmf_women.values, alpha=0.5, label='Women', color='pink')
plt.bar(pmf_men.index, pmf_men.values, alpha=0.5, label='Men', color='blue')
plt.title('PMF of Sleep Hours: Women vs. Men')
plt.xlabel('Sleep Hours')
plt.ylabel('Probability')
plt.legend()
plt.show()


# Compare PMFs Statistically
from scipy.stats import entropy

# Align the PMFs to ensure they have the same index
common_index = pmf_women.index.union(pmf_men.index)
pmf_women_aligned = pmf_women.reindex(common_index, fill_value=0)
pmf_men_aligned = pmf_men.reindex(common_index, fill_value=0)

# Calculate KL Divergence
kl_divergence = entropy(pmf_women_aligned, pmf_men_aligned)
print(f'KL Divergence (Women vs. Men): {kl_divergence:.2f}')
```

## CDF Sleep Hours 
- What percentage of users sleep 6 hours or less?
- How does this compare to the overall distribution of sleep hours?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sort the Sleep Hours data
sorted_sleep = np.sort(df['Sleep Hours'])

# Calculate the CDF
cdf = np.arange(1, len(sorted_sleep) + 1) / len(sorted_sleep)

# Find the proportion of users who sleep 6 hours or less
proportion_6_hours_or_less = np.sum(sorted_sleep <= 6) / len(sorted_sleep)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_sleep, cdf, marker='.', linestyle='none', color='blue', label='CDF of Sleep Hours')

# Highlight the point where Sleep Hours = 6
plt.axvline(x=6, color='red', linestyle='--', label='6 Hours')
plt.axhline(y=proportion_6_hours_or_less, color='green', linestyle='--', label=f'{proportion_6_hours_or_less:.1%} of Users')

# Add labels and title
plt.title('CDF of Sleep Hours (Highlighting 6 Hours)')
plt.xlabel('Sleep Hours')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Analytical Distribution Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Calculate mean and standard deviation of Sleep Hours
mean_sleep = df['Sleep Hours'].mean()
std_sleep = df['Sleep Hours'].std()

# Generate a range of values for the theoretical distribution
x = np.linspace(df['Sleep Hours'].min(), df['Sleep Hours'].max(), 1000)

# Calculate the PDF of the Normal Distribution
pdf = norm.pdf(x, mean_sleep, std_sleep)

# Plot the empirical data (histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Sleep Hours'], bins=20, kde=False, stat='density', color='blue', label='Empirical Data')

# Plot the theoretical Normal Distribution
plt.plot(x, pdf, color='red', label='Theoretical Normal Distribution')

# Add labels and title
plt.title('Sleep Hours: Empirical Data vs. Theoretical Normal Distribution')
plt.xlabel('Sleep Hours')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()


# Quantify the Fit
from scipy.stats import kstest

# Perform the KS test
ks_statistic, p_value = kstest(df['Sleep Hours'], 'norm', args=(mean_sleep, std_sleep))
print(f'KS Statistic: {ks_statistic:.4f}')
print(f'P-value: {p_value:.4f}')
```

## Hypothesis testing
Null Hypothesis (H): Sleep duration before a workout has no effect on calories burned, workout duration, or post-workout mood.
Alternative Hypothesis (H): Sleep duration before a workout influences calories burned, workout duration, and post-workout mood.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, f_oneway
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Check for missing values
print(df.isnull().sum())

# Convert post-workout mood to numerical codes
df['Mood After Workout'] = df['Mood After Workout'].map({'Energized': 2, 'Neutral': 1, 'Fatigued': 0})

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Sleep Hours'], y=df['Calories Burned'], scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
plt.title('Sleep Hours vs. Calories Burned')
plt.xlabel('Sleep Hours')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()

# Calculate Pearson's correlation
correlation, p_value = pearsonr(df['Sleep Hours'], df['Calories Burned'])
print(f"Pearson's Correlation: {correlation:.2f}")
print(f'P-value: {p_value:.4f}')

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Sleep Hours'], y=df['Workout Duration (mins)'], scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
plt.title('Sleep Hours vs. Workout Duration')
plt.xlabel('Sleep Hours')
plt.ylabel('Workout Duration (mins)')
plt.grid(True)
plt.show()

# Calculate Pearson's correlation
correlation, p_value = pearsonr(df['Sleep Hours'], df['Workout Duration (mins)'])
print(f"Pearson's Correlation: {correlation:.2f}")
print(f'P-value: {p_value:.4f}')

# Box plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Mood After Workout'], y=df['Sleep Hours'], palette='Set2')
plt.title('Sleep Hours vs. Post-Workout Mood')
plt.xlabel('Post-Workout Mood')
plt.ylabel('Sleep Hours')
plt.xticks(ticks=[0, 1, 2], labels=['Fatigued', 'Neutral', 'Energized'])
plt.grid(True)
plt.show()

# Perform ANOVA
grouped_data = [df[df['Mood After Workout'] == mood]['Sleep Hours'] for mood in [0, 1, 2]]
f_statistic, p_value = f_oneway(*grouped_data)
print(f'ANOVA F-statistic: {f_statistic:.2f}')
print(f'P-value: {p-value:.4f}')

```

## Regression Analysis
- Null Hypothesis (H): Sleep duration has no effect on calories burned, workout duration, or post-workout mood after controlling for age.
- Alternative Hypothesis (H): Sleep duration has a significant effect on calories burned, workout duration, and post-workout mood after controlling for age.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load the dataset
df = pd.read_csv('workout_fitness_tracker_data.csv')

# Clean the column names
df.columns = df.columns.str.replace(' ', '_')

# Convert Post-Workout Mood to numerical codes
mood_mapping = {'Energized': 2, 'Neutral': 1, 'Fatigued': 0}
df['Mood_After_Workout'] = df['Mood_After_Workout'].map(mood_mapping)

# Check for missing values in the columns used in the models
print("Missing values in each column:")
print(df[['Calories_Burned', 'Sleep_Hours', 'Age', 'Mood_After_Workout']].isnull().sum())

# Drop rows with missing values in the relevant columns
df = df.dropna(subset=['Calories_Burned', 'Sleep_Hours', 'Age', 'Mood_After_Workout'])

# Check for zero-size arrays or columns with only NaN values
for col in ['Calories_Burned', 'Sleep_Hours', 'Age', 'Mood_After_Workout']:
    if df[col].empty or df[col].isnull().all():
        raise ValueError(f"Column '{col}' is empty or contains only NaN values.")

# Add a constant to the independent variables
df = sm.add_constant(df)

# Rename the columns to remove parentheses
df.columns = df.columns.str.replace('(', '').str.replace(')', '')

# Fit the regression models
# Model 1: Calories Burned
model_calories = ols('Calories_Burned ~ Sleep_Hours + Age', data=df).fit()
print("Calories Burned Model Summary:")
print(model_calories.summary())

# Model 2: Workout Duration
model_duration = ols('Workout_Duration_mins ~ Sleep_Hours + Age', data=df).fit()
print("Workout Duration Model Summary:")
print(model_duration.summary())

# Model 3: Mood After Workout
model_mood = ols('Mood_After_Workout ~ Sleep_Hours + Age', data=df).fit()
print("Mood After Workout Model Summary:")
print(model_mood.summary())
```
